```python
from IPython.terminal.shortcuts.auto_suggest import accept
from dotenv import load_dotenv
import os
import requests
import json

load_dotenv()

headers = {
    "accept": "application/json",
    "wanted-client-id": os.getenv("WANTED-CLIENT-ID"),
    "wanted-client-secret": os.getenv("WANTED-CLIENT-SECRET"),
}
params = {
    "query": "AI 엔지니어",
    "limit": "20",
    "offset": "0",
}

basic_url = "https://openapi.wanted.jobs/v1"
```

```python
response = requests.get("https://openapi.wanted.jobs/v1/search/position", headers=headers, params=params)
```

```python
print(response.status_code)
print(json.dumps(response.json(), indent=2))
```

```python
print(json.dumps(response.json(), indent=2))
data = response.json()
```

```python
import pandas as pd

recruit_list = []

for item in data['positions']:
    recruit_id = item['id']
    recruit_position = item['position']
    recruit_url = item['url']
    recruit_company_id = item['company']['id']
    recruit_company_name = item['company']['name']
    recruit_list.append({
        'recruit_id': recruit_id,
        'recruit_position': recruit_position,
        'recruit_url': recruit_url,
        'recruit_company_name': recruit_company_name,
        'recruit_company_id': recruit_company_id
    })

df = pd.DataFrame(recruit_list, index=None)
print(os.getcwd())
df.to_csv("./wanted_recruit_list_engineer.csv", index=False)
```

```python
id_list=[] # parent tag id 518 # 파이선 899
title_list = []
dict = {}
for tag in data['data']:
    for subtag in tag['sub_tags']:
        if subtag['parent_id'] == 518:
            id_list.append(subtag['id'])
            title_list.append(subtag['title'])
            dict["id"] = id_list
            dict["title"] = title_list

df = pd.DataFrame(dict)
df

```

```python
params = {
    "category_tag" :518,
    "subcategory_tags":899

}
response = requests.get("https://openapi.wanted.jobs/v2/jobs", headers=headers, params=params)
```

```python
response.status_code
```

```python
data = response.json()
```

```python
print(json.dumps(data, indent=2))
```

```python
import time
page_ids =[]
title = []
company_names = []
full_locations = []
skills = []
employments = []
links = []
main_tasks = []
preferred = []
requirements = []
hire_rounds = []
for page in data:
    page_id =page['id']
    name = page['name']
    company_name = page['company']['name']
    full_location = page['address']['full_location']
    skills1 = ""
    for elem in page['category_tags']['child_tags']:
        skills1 += elem['title'] + ", "
    employment = page['employment_type']
    link = page['url']

    page_ids.append(page_id)
    title.append(name)
    company_names.append(company_name)
    full_locations.append(full_location)
    skills.append(skills1)
    employments.append(employment)
    links.append(link)
    response = requests.get(f"https://openapi.wanted.jobs/v1/jobs/{page_id}", headers=headers)
    tomak = response.json()
    main_task = tomak['detail']['main_tasks']
    requirement = tomak['detail']['requirements']
    pref = tomak['detail']['preferred_points']
    hire_round = tomak['detail']['hire_rounds']

    main_tasks.append(main_task)
    requirements.append(requirement)
    preferred.append(pref)
    hire_rounds.append(hire_round)

    time.sleep(2)

dict = {
    "page_id": page_ids,
    "title": title,
    "company_name": company_names,
    "full_location": full_locations,
    "skills": skills,
    "employment": employments,
    "link": links
    ,"main_tasks": main_tasks
    ,"requirements": requirements
    ,"preferred": preferred
    ,"hire_rounds": hire_rounds
}

```

```python
df = pd.DataFrame(dict)
df.to_csv("../data/wanted_recruit_complete.csv", index=False)
```
